```python
import requests
from bs4 import BeautifulSoup

def get_events_puerto_rico():
    # URL de búsqueda en Bing u otro motor
    url = "https://www.bing.com/search?q=events+for+tourists+in+Puerto+Rico+2025"
    
    # Realiza la solicitud
    response = requests.get(url)
    if response.status_code != 200:
        print("Error al obtener los resultados.")
        return []

    # Parsear el contenido de la página
    soup = BeautifulSoup(response.content, 'html.parser')

    # Buscar resultados (ajusta según el contenido del motor de búsqueda)
    events = []
    for result in soup.find_all('li', class_='b_algo'):
        title = result.find('h2').text
        link = result.find('a')['href']
        description = result.find('p').text if result.find('p') else 'No description available.'
        
        events.append({
            "title": title,
            "link": link,
            "description": description
        })

    return events

# Ejecutar la función y mostrar los resultados

events = get_events_puerto_rico()
for i, event in enumerate(events):
    print(f"{i+1}. {event['title']}")
    print(f"   Link: {event['link']}")
    print(f"   Description: {event['description']}\n")
```
